
# PDF and Word Documents – Questions & Answers

## 1. In what modes should PdfFileReader() and PdfFileWriter() file objects be opened?
**Answer:**  
- `PdfFileReader()` → opened in **read-binary mode (`'rb'`)**  
- `PdfFileWriter()` → opened in **write-binary mode (`'wb'`)**

**Answer:**  
You must decrypt it:
```python
pdfReader.decrypt('swordfish')
```

## 5. What methods do you use to rotate a page?
**Answer:**  
- `rotateClockwise()`  
- `rotateCounterClockwise()`  

---

## 6. What is the difference between a Run object and a Paragraph object?
**Answer:**  
- **Paragraph** → contains a full paragraph of text  
- **Run** → contains text with the same style within a paragraph  

---

## 7. How do you obtain a list of Paragraph objects from a Document object named doc?
**Answer:**  
```python
doc.paragraphs
```

---

## 8. What type of object has bold, underline, italic, strike, and outline variables?
**Answer:**  
A **Run object**.

---

## 9. What is the difference between False, True, and None for the bold variable?
**Answer:**  
- `True` → text is bold  
- `False` → text is not bold  
- `None` → inherits style from parent  

---

## 10. How do you create a Document object for a new Word document?

```python
from docx import Document
doc = Document()
```

---

## 11. How do you add a paragraph with the text 'Hello, there!' to a Document object named doc?
**Answer:**  
```python
doc.add_paragraph('Hello, there!')
```

---

## 12. What integers represent the heading levels in Word documents?
**Answer:**  
Heading levels are represented by integers **1 to 9**.
