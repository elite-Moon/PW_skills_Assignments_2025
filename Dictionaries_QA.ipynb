
# Dictionaries – Questions & Answers

## 1. What does an empty dictionary's code look like?
**Answer:**  
```python
{}
```

---

## 2. What is the value of a dictionary value with the key 'foo' and the value 42?
**Answer:**  
```python
{'foo': 42}
```

---

## 3. What is the most significant distinction between a dictionary and a list?
**Answer:**  
- A **list** is an ordered collection accessed by index positions.  
- A **dictionary** is an unordered collection accessed by key–value pairs.

---

## 4. What happens if you try to access spam['foo'] if spam is {'bar': 100}?
**Answer:**  
Python raises a **KeyError** because the key `'foo'` does not exist.

---

## 5. Difference between 'cat' in spam and 'cat' in spam.keys()?
**Answer:**  
There is **no difference**.  
Both check whether `'cat'` exists as a **key** in the dictionary.

---

## 6. Difference between 'cat' in spam and 'cat' in spam.values()?
**Answer:**  
- `'cat' in spam` checks **keys**  
- `'cat' in spam.values()` checks **values**

---

## 7. What is a shortcut for the following code?
```python
if 'color' not in spam:
    spam['color'] = 'black'
```

**Answer:**  
```python
spam.setdefault('color', 'black')
```

---

## 8. How do you "pretty print" dictionary values?
**Answer:**  
Use the `pprint` module and the `pprint()` function.

**Example:**
```python
import pprint
pprint.pprint(spam)
```
