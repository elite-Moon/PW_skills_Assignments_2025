
# Lists, Tuples, and Copying – Questions & Answers

## 1. What exactly is []?
**Answer:**  
`[]` represents an empty list in Python.

---

## 2. In a list stored in spam, how would you assign 'hello' as the third value?
**Answer:**  
```python
spam = [2, 4, 6, 8, 10]
spam[2] = 'hello'
```

---

Assume:
```python
spam = ['a', 'b', 'c', 'd']
```

## 3. What is the value of spam[int(int('3' * 2) / 11)]?
**Answer:**  
`'3' * 2` → `'33'`  
`int('33') / 11` → `3`  
`spam[3]` → `'d'`

---

## 4. What is the value of spam[-1]?
**Answer:**  
`'d'`

---

## 5. What is the value of spam[:2]?
**Answer:**  
`['a', 'b']`

---

Assume:
```python
bacon = [3.14, 'cat', 11, 'cat', True]
```

## 6. What is the value of bacon.index('cat')?
**Answer:**  
`1`

---

## 7. How does bacon.append(99) change the list?
**Answer:**  
It adds `99` to the end of the list.

Result:
```python
[3.14, 'cat', 11, 'cat', True, 99]
```

---

## 8. How does bacon.remove('cat') change the list?
**Answer:**  
It removes the **first occurrence** of `'cat'`.

Result:
```python
[3.14, 11, 'cat', True]
```

---

## 9. What are the list concatenation and replication operators?
**Answer:**  
- Concatenation: `+`  
- Replication: `*`

---

## 10. Difference between append() and insert()?
**Answer:**  
- append() adds an item at the end  
- insert() adds an item at a specific index

---

## 11. Two methods for removing items from a list?
**Answer:**  
- remove()  
- pop()

---

## 12. How are list values and string values identical?
**Answer:**  
Both are ordered sequences, support indexing, slicing, and can be used in loops.

---

## 13. Difference between tuples and lists?
**Answer:**  
- Lists are mutable  
- Tuples are immutable

---

## 14. How do you type a tuple with only the integer 42?
**Answer:**  
```python
(42,)
```

---

## 15. How do you convert between list and tuple?
**Answer:**  
- List to tuple: `tuple(list_value)`  
- Tuple to list: `list(tuple_value)`

---

## 16. What do variables that "contain" list values actually contain?
**Answer:**  
They contain **references** to list objects in memory.

---

## 17. Difference between copy.copy() and copy.deepcopy()?
**Answer:**  
- copy.copy() makes a shallow copy  
- copy.deepcopy() makes a deep copy (copies nested objects too)
