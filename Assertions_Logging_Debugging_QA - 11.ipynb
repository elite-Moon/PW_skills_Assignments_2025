
# Assertions, Logging, and Debugging – Questions & Answers

## 1. Create an assert statement that throws an AssertionError if the variable spam is a negative integer.
**Answer:**
```python
assert spam >= 0, "spam must not be negative"
```

---

## 2. Assert that eggs and bacon are not the same string (case-insensitive).
**Answer:**
```python
assert eggs.lower() != bacon.lower(), "eggs and bacon should not be the same"
```

---

## 3. Create an assert statement that throws an AssertionError every time.
**Answer:**
```python
assert False, "This assertion always fails"
```

---

## 4. What are the two lines required to call logging.debug()?
**Answer:**
```python
import logging
logging.basicConfig(level=logging.DEBUG)
```

---

## 5. What two lines are required to log messages to programLog.txt?
**Answer:**
```python
import logging
logging.basicConfig(filename='programLog.txt', level=logging.DEBUG)
```

---

## 6. What are the five levels of logging?
**Answer:**  
1. DEBUG  
2. INFO  
3. WARNING  
4. ERROR  
5. CRITICAL  

---

## 7. What line of code disables all logging messages?
**Answer:**
```python
logging.disable(logging.CRITICAL)
```

---

## 8. Why is logging better than print()?
**Answer:**  
Logging can be turned on/off, saved to files, filtered by severity level, and does not need to be removed after debugging.

---

## 9. Difference between Step Over, Step In, and Step Out?
**Answer:**  
- Step Over → executes the current line without entering functions  
- Step In → enters the called function  
- Step Out → finishes the current function and returns  

---

## 10. After clicking Continue, when does the debugger stop?
**Answer:**  
The debugger stops at the next breakpoint or when the program ends.

---

## 11. What is a breakpoint?
**Answer:**  
A breakpoint is a marker set on a line of code where program execution pauses during debugging.
