## 1. Why are functions advantageous to have in your programs?
**Answer:**  
Functions help reduce repetition, improve readability, make debugging easier, and allow code reuse.

---

## 2. When does the code in a function run?
**Answer:**  
The code runs only when the function is called.

---

## 3. What statement creates a function?
**Answer:**  
The `def` statement creates a function.

---

## 4. Difference between a function and a function call?
**Answer:**  
A function defines code; a function call executes it.

---

## 5. Global and local scopes?
**Answer:**  
One global scope and one local scope per function call.

---

## 6. What happens to local variables after function returns?
**Answer:**  
They are destroyed.

---

## 7. What is a return value?
**Answer:**  
It is the value a function sends back; yes, it can be used in expressions.

---

## 8. Return value without return statement?
**Answer:**  
`None`

---

## 9. How to refer to a global variable inside a function?
**Answer:**  
Use the `global` keyword.

---

## 10. Data type of None?
**Answer:**  
`NoneType`
