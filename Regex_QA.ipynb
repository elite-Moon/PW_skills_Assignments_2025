
# Regular Expressions (Regex) – Questions & Answers

## 1. What is the name of the feature responsible for generating Regex objects?
**Answer:**  
The `re.compile()` function is responsible for generating Regex objects.

---

## 2. Why do raw strings often appear in Regex objects?
**Answer:**  
Raw strings (prefixed with `r`) prevent Python from interpreting backslashes as escape characters.

---

## 3. What is the return value of the search() method?
**Answer:**  
It returns a **Match object** if the pattern is found, otherwise `None`.

---

## 4. From a Match object, how do you get the actual strings that match the pattern?
**Answer:**  
Use the `group()` method.

---

## 5. In regex r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what do the groups cover?
**Answer:**  
- Group 0 → Entire match  
- Group 1 → First three digits  
- Group 2 → Last seven digits  

---

## 6. How do you match literal parentheses and periods in regex?
**Answer:**  
By escaping them with backslashes:
```regex
\( \) \.
```

---

## 7. Why does findall() return strings or tuples?
**Answer:**  
- No groups → list of strings  
- Groups present → list of tuples  

---

## 8. What does the | character mean?
**Answer:**  
It means **OR**.

---

## 9. What does the . character stand for?
**Answer:**  
It matches any character except newline.

---

## 10. Difference between + and *?
**Answer:**  
- `+` → one or more  
- `*` → zero or more  

---

## 11. Difference between {4} and {4,5}?
**Answer:**  
- `{4}` → exactly 4  
- `{4,5}` → 4 or 5  

---

## 12. Meaning of \d, \w, \s?
**Answer:**  
- `\d` → digit  
- `\w` → letter, digit, underscore  
- `\s` → whitespace  

---

## 13. Meaning of \D, \W, \S?
**Answer:**  
Opposites of above:
- `\D` → non-digit  
- `\W` → non-word  
- `\S` → non-whitespace  

---

## 14. Difference between .*? and .*?
**Answer:**  
- `.*` → greedy  
- `.*?` → non-greedy  

---

## 15. Regex for numbers and lowercase letters?
**Answer:**  
```regex
[a-z0-9]
```

---

## 16. How to make regex case-insensitive?
**Answer:**  
Pass `re.IGNORECASE` to `re.compile()`.

---

## 17. What does . match normally and with re.DOTALL?
**Answer:**  
- Normally → any character except newline  
- With DOTALL → includes newline  

---

## 18. Output of substitution?
```python
import re
numRegex = re.compile(r'\d+')
numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen')
```
**Answer:**  
`X drummers, X pipers, five rings, X hen`

---

## 19. What does re.VERBOSE allow?
**Answer:**  
It allows whitespace and comments in regex for readability.

---

## 20. Regex for comma-separated numbers?
**Answer:**  
```regex
^\d{1,3}(,\d{3})*$
```

---

## 21. Regex for full name ending with Watanabe?
**Answer:**  
```regex
^[A-Z][a-zA-Z]*\sWatanabe$
```

---

## 22. Regex for specific sentences (case-insensitive)?
**Answer:**  
```regex
^(Alice|Bob|Carol)\s(eats|pets|throws)\s(apples|cats|baseballs)\.$
```
Use with `re.IGNORECASE`.
