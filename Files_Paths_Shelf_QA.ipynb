
# Files, Paths, and Shelves – Questions & Answers

## 1. To what does a relative path refer?
**Answer:**  
A relative path refers to a file or folder location **relative to the current working directory**, not from the root of the file system.

---

## 2. What does an absolute path start with on your operating system?
**Answer:**  
- On **Windows**, an absolute path starts with a drive letter, e.g. `C:\`  
- On **Linux/macOS**, it starts with `/`

---

## 3. What do the functions os.getcwd() and os.chdir() do?
**Answer:**  
- `os.getcwd()` → returns the current working directory  
- `os.chdir(path)` → changes the current working directory to `path`

---

## 4. What are the . and .. folders?
**Answer:**  
- `.` → current directory  
- `..` → parent directory  

---

## 5. In `C:\bacon\eggs\spam.txt`, which part is the dir name and base name?
**Answer:**  
- Directory name → `C:\bacon\eggs`  
- Base name → `spam.txt`

---

## 6. What are the three mode arguments for the open() function?
**Answer:**  
- `'r'` → read mode  
- `'w'` → write mode  
- `'a'` → append mode  

---

## 7. What happens if an existing file is opened in write mode?
**Answer:**  
The file’s contents are **erased** and replaced with new content.

---

## 8. Difference between read() and readlines()?
**Answer:**  
- `read()` → returns the entire file as a single string  
- `readlines()` → returns a list of strings, one per line  

---

## 9. What data structure does a shelf value resemble?
**Answer:**  
A shelf value resembles a **dictionary** with key–value pairs.
