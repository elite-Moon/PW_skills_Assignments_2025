
# PyInputPlus – Questions & Answers

## 1. Is the Python Standard Library included with PyInputPlus?
**Answer:**  
No. PyInputPlus is a **third-party module** and is not included in the Python Standard Library.

---

## 2. Why is PyInputPlus commonly imported as `import pyinputplus as pypi`?
**Answer:**  
To save typing and make the code shorter and more readable.

---

## 3. How do you distinguish between inputInt() and inputFloat()?
**Answer:**  
- `inputInt()` → accepts only integer values  
- `inputFloat()` → accepts decimal (floating-point) values  

---

## 4. Using PyInputPlus, how do you ensure the user enters a whole number between 0 and 99?
**Answer:**  
```python
import pyinputplus as pypi
pypi.inputInt(min=0, max=99)
```

---

## 5. What is transferred to the keyword arguments allowRegexes and blockRegexes?
**Answer:**  
They accept a **list of regular expression strings** that are allowed or blocked as valid input.

---

## 6. If a blank input is entered three times, what does inputStr(limit=3) do?
**Answer:**  
It raises a **RetryLimitException** after three failed attempts.

---

## 7. If blank input is entered three times, what does inputStr(limit=3, default='hello') do?
**Answer:**  
It returns the default value `'hello'` instead of raising an exception.
